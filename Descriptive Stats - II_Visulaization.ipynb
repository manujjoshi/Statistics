### `Pie chart`

```python
value = [154,411,34,141,25,256]
labels = ["china","india","USA","Indonesai","Brazil","Pakistan"]
plt.figure(figsize=(24,8))
plt.pie(value,labels=labels,autopct="%0.2f%%")
plt.legend(labels,loc=1)
plt.show()
```

### `Scatter plot`

```python
np.random.seed(10)
height = np.random.normal(140,180,size=100)
weight = np.random.normal(60,80,size=100)
plt.scatter(weight,height,c="blue",marker=".")
plt.xlabel("Weight",fontsize="15")
plt.ylabel("Height",fontsize="15")
plt.title("Height vs Weight",fontsize=25)
plt.show()
```

```python
np.random.seed(10)

height_m = np.random.normal(140,20,size=100)
weight_m = np.random.normal(60,80,size=100)


height_f = np.random.normal(80,20,size=100)
weight_f = np.random.normal(40,80,size=100)


plt.scatter(weight_m,height_m,c="blue",marker=".")
plt.scatter(weight_f,height_f,c="green",marker="*")
plt.xlabel("Weight",fontsize="15")
plt.ylabel("Height",fontsize="15")
plt.title("Height vs Weight",fontsize=25)
plt.show()
```

### `Barplot`

```python
# Simple bar
value = [154,411,34,141,25,256]
labels = ["china","india","USA","Indonesai","Brazil","Pakistan"]
plt.figure(figsize=(24,8))
plt.bar(labels,value,color="skyblue",edgecolor="red")
plt.xlabel("Country",fontsize="15")
plt.ylabel("Population",fontsize="15")
plt.legend(labels,loc=1)
plt.show()
```

```python
# horizontal bar
value = [154,411,34,141,25,256]
labels = ["china","india","USA","Indonesai","Brazil","Pakistan"]
plt.figure(figsize=(24,8))
plt.barh(labels,value,color="skyblue",edgecolor="red")
plt.xlabel("Country",fontsize="15")
plt.ylabel("Population",fontsize="15")
plt.legend(labels,loc=1)
plt.show()
```

```python
# Stacked bar
urban_value = [154,411,34,141,25,256]
rural_value = [54,41,3,11,25,26]
labels = ["china","india","USA","Indonesai","Brazil","Pakistan"]
plt.figure(figsize=(24,8))
plt.bar(labels,urban_value,color="skyblue",edgecolor="red")
plt.bar(labels,rural_value,color="red",edgecolor="red")
plt.xlabel("Country",fontsize="15")
plt.ylabel("Population",fontsize="15")
plt.legend(["urban","rural"],loc=1)
plt.show()
```

```python
# Grouped bar
urban_value = [154,411,34,141,25,256]
rural_value = [54,41,3,11,25,26]
labels = ["china","india","USA","Indonesai","Brazil","Pakistan"]
plt.figure(figsize=(24,8))
plt.bar((0,1,2,3,4,5),urban_value,color="#C7190B",width=0.4,edgecolor="#B0C70A")
plt.bar((0.4,1.4,2.4,3.4,4.4,5.4),rural_value,color="#B0C70A",width=0.4,edgecolor="#C7190B")
plt.xticks([0,1,2,3,4,5],labels)
plt.xlabel("Country",fontsize="15")
plt.ylabel("Population",fontsize="15")
plt.legend(["urban","rural"],loc=1)
plt.show()
```

### `subplot`

```python
plt.figure(figsize=(15,10))
plt.subplot(221)
sns.countplot(x='day',data=tips_df)

plt.subplot(222)
sns.heatmap(tips_df.corr(),annot=True)

plt.subplot(223)
sns.boxplot(x='day',y='tip',data=tips_df)

plt.subplot(224)
sns.scatterplot(x='total_bill',y='tip',data=tips_df)

plt.show()
```
